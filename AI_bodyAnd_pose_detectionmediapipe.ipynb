```python
import mediapipe as mp
import cv2
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

### Get realtime Webcam feed

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret , frame = cap.read()
    print(frame)
    cv2.imshow("web camp, ", frame)
    
    if cv2.waitKey(10) & 0xFF==ord('q'):
        break
cap.release()
# if you call destroyWindow with cv2 then it required some title argument to finish with this argument e.g destroyWindow('hhahaha')
# otherwise it give error---> missing required argument
# if you call destroyAllWindows() then nothing required it will finish automatically
# cv2.destroyWindow('hahhhaha')
cv2.destroyAllWindows()
    
```

## Make Detection from feed
#### 1: detection facial landmark   
#### 2: detect hand poses
#### 3: detect body poses

```python
cap = cv2.VideoCapture(0)

# initiate holistic model
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as hm:
    
    while cap.isOpened():
        ret , frame = cap.read()
        
        # recolor feed/frame b/z webcamp have frame with bgr color
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # MAKE DETECTION
        result = hm.process(image)
#         print(result.face_landmarks)
        
        
        #face_landmarks, pose_landmarks, right_hand_landmarks, left_hand_marks
        
        # reconvert image to bgr b/z webcamp love with bgr color
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # draw face_landmarks
        mp_drawing.draw_landmarks(image, result.face_landmarks, mp_holistic.FACEMESH_TESSELATION,
                                 mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=2 ),
                                 mp_drawing.DrawingSpec(color=(0,255,0), thickness=3, circle_radius=1 )
)
        
        
        # draw right_landmarks
        mp_drawing.draw_landmarks(image, result.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(255,0,0), thickness=3, circle_radius=2 ),
                                 mp_drawing.DrawingSpec(color=(0,255,0), thickness=3, circle_radius=2)
)
        
        # draw left_landmarks
        mp_drawing.draw_landmarks(image, result.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(255,0,0), thickness=4, circle_radius=2 ),
                                 mp_drawing.DrawingSpec(color=(0,255,0), thickness=4, circle_radius=2 )
)
        
#         draw pose landmarks
        mp_drawing.draw_landmarks(image, result.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(255,0,0), thickness=1, circle_radius=3 ),
                                 mp_drawing.DrawingSpec(color=(0,255,0), thickness=1, circle_radius=3 )
                                  
)
        cv2.imshow('my webcam: ', image)
        
        
    
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```
